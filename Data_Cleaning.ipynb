```python
def plot_histogram_from_csv(file_var, column_name, bins=10):
    """
    Reads a column from a CSV file and plots its histogram.

    Parameters:
        column_name (str): Name of the column to plot.

    Returns:
        None
    """
    # Check if the column exists in the CSV
    if column_name not in file_var.columns:
        print(f"Column '{column_name}' not found in the CSV file.")
        return

    # Extract the column data
    column_data = file_var[column_name]

    # Ensure the column contains numeric data
    if not pd.api.types.is_numeric_dtype(column_data):
        print(f"Column '{column_name}' is not numeric and cannot be used for a histogram.")
        return

    # Plot the histogram
    plt.figure(figsize=(8, 6))
    plt.hist(column_data.dropna(), bins=bins, color='blue', edgecolor='black', alpha=0.7)
    plt.title(f"Histogram of {column_name}")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
data = []

def plot_boxplot_from_csv(file_var, column_name, remove_outliers=False):
    """
    Reads a column from a CSV file and plots its box plot.

    Parameters:
        column_name (str): Name of the column to plot.

    Returns:
        None
    """
    # Check if the column exists in the CSV
    if column_name not in file_var.columns:
        print(f"Column '{column_name}' not found in the CSV file.")
        return

    # Extract the column data
    column_data = file_var[column_name]

    # Ensure the column contains numeric data
    if not pd.api.types.is_numeric_dtype(column_data):
        print(f"Column '{column_name}' is not numeric and cannot be used for a box plot.")
        return

    # Calculate statistics for outlier detection
    q1 = column_data.quantile(0.25)
    q3 = column_data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    print(f"Outlier range for '{column_name}':")
    print(f"  Minimum outlier threshold: {lower_bound}")
    print(f"  Maximum outlier threshold: {upper_bound}")

    # Remove outliers if specified
    if remove_outliers:
        original_length = len(column_data)
        updated_data = column_data[(column_data >= lower_bound) & (column_data <= upper_bound)]
        updated_length = len(column_data)
        data.append(column_data)
        print(f"Outliers removed: {original_length - updated_length} rows")

        #Save the updated data back to the CSV file
        #column_data.to_csv(file_var, index=False)
        #print(f"CSV file '{file_var}' updated after removing outliers.")
        
    # Plot the box plot
    plt.figure(figsize=(8, 6))
    plt.boxplot(column_data.dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'), flierprops=dict(marker='o', color='red', alpha=0.6))
    plt.title(f"Box Plot of {column_name}")
    plt.xlabel(column_name)
    plt.grid(axis='x', alpha=0.75)
    plt.show()
    print(updated_data);
```

```python
for col_name in unique_realAdultData.columns:
    current_column = col_name;
    plot_boxplot_from_csv(unique_realAdultData, current_column, True);
```
